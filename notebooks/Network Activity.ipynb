```python
import os, sys
from os.path import expanduser
import matplotlib.pyplot as plt
%matplotlib inline

user_home = expanduser("~")
neuron_home = '%s/bin/nrnpython/lib/python' % user_home
model_home = '%s/src/model' % user_home
dentate_home = '%s/dentate' % model_home
results_prefix = os.getenv('DG_RESULTS_DIR', '%s/results' % dentate_home)
sys.path.append(neuron_home) 
sys.path.append(model_home) 

import dentate
from dentate import utils, plot
from dentate.env import Env
```

```python
spike_events_path = '%s/Full_Scale_GC_Exc_Sat_LNN_Diag_10090593.bw/dentatenet_Full_Scale_GC_Exc_Sat_LNN_results.h5' % results_prefix
trajectory_path = '%s/datasets/DG_PP_spiketrains_A_20190512.h5' % dentate_home

```

```python

font_size = 32
populations=['IS', 'MOPP', 'NGFC', 'HCC', 'HC', 'BC', 'AAC', 'MC', 'GC']
    
time_range = [2000.0, 4000.0]
spike_hist_bin = 5.0
fig = plot.plot_spike_raster (spike_events_path, 'Spike Events', include=populations, time_range=time_range, 
                              time_variable='t', pop_rates=True, spike_hist='subplot', 
                              max_spikes=None, spike_hist_bin=spike_hist_bin, 
                              fontSize=font_size, labels='yticks', saveFig=True, figFormat='png')
```

```python
populations=['GC', 'MC', 'BC', 'HC', 'HCC', 'NGFC']

fig = plot.plot_spike_histogram (spike_events_path, 'Spike Events', include=populations, time_variable='t',
                                 time_range=time_range, pop_rates=False, bin_size=spike_hist_bin,
                                 smooth=0, quantity='count', fontSize=font_size, overlay=False, graph_type='bar', 
                                 figSize=(15,15), saveFig=True, figFormat='svg')
```
